```python
import random

list_random = [random.randint(0, 500) for _ in range(50, 500, 10)]
print(list_random)
```

```python
swap_insertion = 0
swap_selection = 0
swap_quick = 0
comp_insertion = 0
comp_selection = 0
comp_quick = 0
```

```python
def swap_inser(arr, index1, index2):
    # Swaps the elements at index1 and index2 in the input list
    global swap_insertion
    temp = arr[index1]
    arr[index1] = arr[index2]
    arr[index2] = temp
    swap_insertion += 1
```

```python
def swap_selec(arr, index1, index2):
    # Swaps the elements at index1 and index2 in the input list
    global swap_selection
    temp = arr[index1]
    arr[index1] = arr[index2]
    arr[index2] = temp
    swap_selection += 1
```

```python
def swap(arr, index1, index2):
    # Swaps the elements at index1 and index2 in the input list
    global swap_quick
    temp = arr[index1]
    arr[index1] = arr[index2]
    arr[index2] = temp
    swap_quick += 1
```

```python
def selection_sort(arr):
    # Sorts the input list using the selection sort algorithm
    for i in range(len(arr) - 1):
        min_index = find_minimum(arr, i)
        swap_selec(arr, i, min_index)

def find_minimum(arr, start_index):
    # Returns the index of the minimum element in the subarray starting at start_index
    global comp_selection
    min_index = start_index
    for j in range(start_index, len(arr)):
        comp_selection += 1
        if arr[min_index] > arr[j]:
            min_index = j
    return min_index
```

```python
def insertion_sort(arr):
    # Sorts the input list using the insertion sort algorithm
    for i in range(len(arr)):
        pushdown(arr, i)

def pushdown(arr, i):
    # Pushes down the element at index i to its correct position in the sorted subarray
    global comp_insertion
    j = 0
    while j < i:
        comp_insertion += 1
        if arr[i-j] < arr[i-j-1]:
            swap_inser(arr, i-j, i-j-1)
            j += 1
        else:
            break
```

```python
def quick_sort(arr, low, high):
    # Sorts the input list using the quick sort algorithm
    if (high - low) < 1:
        return None
    pivot_index = partition(arr, low, high)
    quick_sort(arr, low, pivot_index - 1)
    quick_sort(arr, pivot_index + 1, high)

def partition(arr, low, high):
    # Partitions the subarray arr[low..high] around a pivot and returns the new pivot index
    global comp_quick
    pivot = arr[low]
    i = low
    j = high + 1
    while i < j - 1:
        comp_quick += 1
        if arr[i + 1] >= pivot:
            swap(arr, i + 1, j - 1)
            j -= 1
        else:
            swap(arr, i, i + 1)
            i += 1
    return i
```

```python
# Initialize lists to store the number of swaps and comparisons for each sorting algorithm for each input size (100, 1000 and 10000)
comp_insertion_list = []
swap_insertion_list = []
comp_selection_list = []
swap_selection_list = []
comp_quick_list = []
swap_quick_list = []
```

```python
# Generate and sort random lists of size 500 using each sorting algorithm and store the number of swaps and comparisons in the corresponding lists.
for _ in range(50):
    
    # Insertion sort
    insertion_sort(list_random[:])
    comp_insertion_list.append(comp_insertion)
    swap_insertion_list.append(swap_insertion)
    
    # Selection sort
    selection_sort(list_random[:])
    comp_selection_list.append(comp_selection)
    swap_selection_list.append(swap_selection)
    
    # Quick sort
    quick_sort(list_random[:], 0 , len(list_random) - 1)
    comp_quick_list.append(comp_quick)
    swap_quick_list.append(swap_quick)
```

```python
import matplotlib.pyplot as plt

# Plot the number of comparisons for each sorting algorithm for input list
plt.plot(['Insertion sort', 'Selection sort', 'Quick sort'], [sum(comp_insertion_list) / 50, sum(comp_selection_list) / 50, sum(comp_quick_list) / 50])
plt.title('Number of comparisons for input list')
plt.show()

# Plot the number of swaps for each sorting algorithm for input list
plt.plot(['Insertion sort', 'Selection sort', 'Quick sort'], [sum(swap_insertion_list) / 50, sum(swap_selection_list) / 50, sum(swap_quick_list) / 50])
plt.title('Number of swaps for input list')
plt.show()

# Plot the number of swaps and comparisions for each sorting algorithm for input list
plt.plot(['Insertion sort', 'Selection sort', 'Quick sort'], [sum(swap_insertion_list) / 50 + sum(comp_insertion_list) / 50, sum(swap_selection_list) / 50 + sum(comp_selection_list) / 50, sum(swap_quick_list) / 50 + sum(comp_quick_list) / 50])
plt.title('Number of swaps + comparision for input list')
plt.show()
```
